## Проверка завершена

Так как нет данных о количестве степов в курсе, предположим, что максимальное число  уникальных открытых степов _`unique_step_id_number`_ в датафрейме _`unique_actions_stats`_ и есть суммарное количество степов в курсе. Будем считать, что это предположение верно, если не менее чем у 2-х пользователей число уникальных пройденных степов равно _`unique_step_id_number`_.

```python
course_total_steps_number = None
unique_step_id_number = unique_actions_stats.unique_steps_discovered.max()
passed_steps_number = unique_actions_stats.unique_steps_discovered
if (passed_steps_number == unique_step_id_number).value_counts()[True] > 2:
    course_total_steps_number = unique_step_id_number
    print(f'Предположение верно. Общее количество степов в курсе: {course_total_steps_number}.')
else:
    print(f'Предположение неверно. Общее количество степов в курсе неизвестно.')
```

```python
%%time
# Рассчитаем перерывы в активности студентов.

gaps_data = events_data.assign(date = events_data.datetime.dt.date)[['user_id', 'date', 'datetime']]\
.drop_duplicates(subset=['user_id', 'date'])\
.groupby('user_id')\
.datetime\
.apply(list)\
.apply(np.diff)\
.apply(np.abs)\
.values

gaps_data = pd.Series(np.concatenate(gaps_data, axis=0)).dt.days
```

```python
def get_datetime_indicators(users_actions_stats,
                            entrance_datetime = 'entrance_datetime',
                            status_getting_datetime = 'status_getting_datetime',
                            user_id = 'user_id',
                            frequency = 'M'):
    '''
    Определяет основные характеристики курса, описывающие его в динамике
    
    users_actions_stats:
    датафрейм с данными о действиях пользователей курса
    
    entrance_datetime = "entrance_datetime" default:
    название колонки, содержащей даты начала обучения студентов в формате datetime
    
    status_getting_datetime = "status_getting_datetime" default:
    название колонки, содержащей даты начала присвоения студентам статуса "in progress", "passed course", 
    "dropped out" в формате datetime
    
    user_id = "user_id" default:
    колонка с идентификаторами пользователей
    
    frequency = 'M' default:
    определяет частоту агрегации данных по времени 'M' (month), 'D' (day) etc.
    '''
    # Не учитываем временной промежуток длинной drop_out_gap от последней даты в датафрейме.
    # Требуется модель для прогнозирования дропнутся или нет эти пользователи.
    course_period = pd.date_range(start = min_events_datetime.strftime('%Y-%m-%d'),
                                  end = (max_events_datetime - timedelta(drop_out_gap)).strftime('%Y-%m-%d'),
                                  freq = frequency)
    # Создаём пустой датафрейм с колонкой в формате datetime
    users_stats = pd.DataFrame({'date':course_period})
         
    for indicator in ['joined', 'passed_course', 'dropped_out']:
        datetime_column = None
        if indicator == 'joined':
            datetime_column = entrance_datetime
            sub_df = users_actions_stats[[datetime_column, user_id]]
        else:
            datetime_column = status_getting_datetime
            sub_df = users_actions_stats\
                    .loc[users_actions_stats.status == indicator.replace('_',' ')]\
                    [[status_getting_datetime, user_id]]
                
        sub_df = sub_df.set_index(datetime_column)
        users_stats = users_stats.merge(sub_df
                                        .resample(frequency)
                                        .count()
                                        .reset_index()
                                        .rename(columns={datetime_column:'date', user_id:indicator}),
                                        how='left',
                                        on='date')
    
    users_stats = users_stats.fillna(0).set_index('date').astype('int')
            
    users_stats['in_progress'] = np.add.accumulate(users_stats.joined
                                                   - users_stats.passed_course
                                                   - users_stats.dropped_out)
    return users_stats
```

```python
# Топ-5 степов, после которых студенты бросали курс 

actions_translation = {'discovered':'открыв',\
                       'passed':'пройдя',\
                       'started_attempt':'попытавшись выполнить',\
                       'viewed':'просмотрев'}

potentially_hard_steps_list = []

for action in normalized_last_action_stats.columns:    
    plt.figure()
    plt.ylabel(f'{action}.\nNormalized value')
    top_5 = normalized_last_action_stats[action]\
    .sort_values(ascending=False)\
    .head(5)
    ax = top_5.plot.bar(ylim=(0,0.12))
    
    for p in ax.patches:
        ax.annotate(str(round(p.get_height(),3)), (p.get_x()+0.5*p.get_width(), p.get_height() * 1.005),
                    va='bottom', ha='center')
    
    ax.text(0.6,0.1,f'total_value = {round(top_5.sum(),3)}', fontsize=14)
    plt.show()
    
    print(f'{round(top_5.sum()*100,1)}% студентов покинули курс {actions_translation[action]} задания с\
user_id = {top_5.index.to_list()}.\n')
    
    
    potentially_hard_steps_list.append(top_5.index.to_list())

potentially_hard_steps_set = set(np.concatenate(potentially_hard_steps_list, axis=0)) 
print(f'Степы со step_id = {potentially_hard_steps_set} могут быть потенциально сложными для студентов.')
```

**Подготовим функцию, которая будет обрабатывать данные датафреймов типа `events_data` и `submission_data`, приводить их к виду аналогичному `users_actions_stats` и `last_session`, определяя статус пользователей по итогу анализа данных за весь период, а также показатели пользователей за `days_to_train` дней с начала их обучения на курсе. При этом
`days_to_train < course_completion_time`. Эти данные будут использоваться для обучения модели, которая будет прогнозировать какие из студентов со статусом *'in_progress'* уйдут, а какие нет по их активности за `days_to_train` дней.** 

**Тем студентам для которых прогноз будет *'dropped_out'* можно попробовать придумать стимулирующие бонусы.**

**Необходимо также определить оптимальное значение _`days_to_train`_ для модели (один из параметров оптимизации).**

```python
class StudentsStatusClassifier:
    def __init__(self, 
                 events_data_path = 'materials/event_data_train.csv', 
                 submissions_data_path = 'materials/submissions_data_train.csv'):
        '''
        '''
        events_data = pd.read_csv(events_data_path)
        events_data['datetime'] = pd.to_datetime(events_data.timestamp, unit='s')
        events_data['date'] = events_data.datetime.dt.date
        
        submissions_data = pd.read_csv(submissions_data_path)
        submissions_data['datetime'] = pd.to_datetime(submissions_data.timestamp, unit='s')
        submissions_data['date'] = submissions_data.datetime.dt.date
        
        self.events_data = events_data
        self.submissions_data = submissions_data
        
        
    def get_drop_out_gap(self,
                         drop_out_gap_quantile = 0.95):
        '''
        Расчет перерывов в активности студентов и определение наиболее часто встречающихся значений.
        '''       
        gaps_data = self.events_data[['user_id', 'date', 'datetime']]\
        .drop_duplicates(subset=['user_id', 'date'])\
        .groupby('user_id')\
        .datetime\
        .apply(list)\
        .apply(np.diff)\
        .apply(np.abs)\
        .values

        gaps_data = pd.Series(np.concatenate(gaps_data, axis=0)).dt.days
        self.drop_out_gap = int(gaps_data.quantile(drop_out_gap_quantile))
        return self.drop_out_gap
    
    
    def get_students_status(self,
                            course_total_steps_number = 198):
        '''
        Определение статуса студентов.
        '''
        # Определение времени последнего действия студентов
        last_action_data = self.events_data\
                          .groupby('user_id', as_index=False)\
                          .agg({'datetime':'max'})\
                          .rename(columns={'datetime':'last_datetime'})

        # Определение показателей активности студентов за весь период мониторинга данных
        total_actions_stats = pd.pivot_table(data=self.events_data,
                                                  values='step_id',
                                                  index='user_id',
                                                  columns='action',
                                                  aggfunc='count',
                                                  fill_value=0)\
                                     .rename(columns={'discovered':'total_discovers',
                                                      'passed':'total_passes',
                                                      'started_attempt':'total_attempts',
                                                      'viewed':'total_views'})\
                                     .reset_index()

        users_status = total_actions_stats[['user_id', 'total_passes']].merge(last_action_data, 
                                                            how='outer', 
                                                            on='user_id')

        # Определение статуса студентов
        users_status\
        .loc[users_status.total_passes == course_total_steps_number, 'status'] = 'passed course' 

        self.max_events_datetime = self.events_data.datetime.max()

        users_status\
        .loc[(~ users_status.status.isin(['passed course'])) &\
             ((self.max_events_datetime - users_status.last_datetime).dt.days > self.drop_out_gap), 'status'] = 'dropped out'

        users_status = users_status\
                       .fillna(value={'status': 'in progress'})\
                       .drop(columns=['total_passes', 'last_datetime'])
        
        self.users_status = users_status
        return self.users_status
    
    
    def get_course_completion_time(self,
                                   course_completion_time_quantile = 0.95):
        '''
        Определение времени, за которое большинство студентов заканчивает курс.
        '''
        # Определение количества уникальных дней, которые студенты потратили на обучение за весь период мониторинга данных.
        unique_days = self.events_data\
                      .groupby('user_id', as_index=False)\
                      .agg({'date':'nunique'})\
                      .rename(columns={'date':'unique_days'})

        unique_days = unique_days.merge(self.users_status[['user_id', 'status']],
                                                  how='outer',
                                                  on='user_id')

        # Определение времени, за которое большинство студентов заканчивает курс.
        passed_course = unique_days.loc[unique_days.status == "passed course"]
        
        self.course_completion_time = int(passed_course.unique_days.quantile(course_completion_time_quantile))
        return self.course_completion_time

    def get_users_actions_stats(self, 
                                days_to_train_):
        self.days_to_train = days_to_train_

        # Определение даты начала обучения студентов
        entrance_datetime = self.events_data\
                                .groupby('user_id')\
                                .agg({'datetime':'min'})\
                                .rename(columns={'datetime':'entrance_datetime'})
        
        users_actions_stats = self.users_status\
                                  .merge(entrance_datetime,
                                         how='outer',
                                         on='user_id')

        # Отбор данных за период days_to_train с момента начала обучения каждого студента.
        events_data_ = self.events_data.merge(entrance_datetime,
                                              how='outer',
                                              on='user_id')
        days_to_train_events_data = events_data_\
        .loc[events_data_.datetime <= events_data_.entrance_datetime + timedelta(self.days_to_train)]
    
        submissions_data_ = self.submissions_data.merge(entrance_datetime,
                                                   how='outer',
                                                   on='user_id')
        days_to_train_submissions_data = submissions_data_\
        .loc[submissions_data_.datetime <= submissions_data_.entrance_datetime + timedelta(self.days_to_train)]

        # Определение количества уникальных дней, которые студенты потратили на обучение за период days_to_train.
        days_to_train_unique_days = days_to_train_events_data\
                                    .groupby('user_id', as_index=False)\
                                    .agg({'date':'nunique'})\
                                    .rename(columns={'date':'unique_days'})
        users_actions_stats = users_actions_stats.merge(days_to_train_unique_days,
                                                        how='outer',
                                                        on='user_id')

        # Определение показателей активности студентов за период days_to_train
        days_to_train_actions_stats = pd.pivot_table(data=days_to_train_events_data,
                                                  values='step_id',
                                                  index='user_id',
                                                  columns='action',
                                                  aggfunc='count',
                                                  fill_value=0)\
                                     .reset_index()

        users_actions_stats = users_actions_stats.merge(days_to_train_actions_stats, 
                                                        how='outer', 
                                                        on='user_id')

        # Определение количества правильных и неправильных попыток выполнения заданий студентами за период days_to_train
        days_to_train_users_submissions = pd.pivot_table(data=days_to_train_submissions_data,
                                                         index='user_id',
                                                         columns='submission_status',
                                                         values='step_id',
                                                         aggfunc='count',
                                                         fill_value=0)\
                                            .rename(columns=lambda x: f'all_{x}_submissions')\
                                            .reset_index()

        users_actions_stats = users_actions_stats.merge(days_to_train_users_submissions,
                                                        how='outer',
                                                        on='user_id')\
                                                 .fillna(0)

        users_actions_stats\
        .loc[users_actions_stats.all_correct_submissions + users_actions_stats.all_wrong_submissions != 0, 'user_success_rate']\
        = round(users_actions_stats.all_correct_submissions / (users_actions_stats.all_correct_submissions
                                                           + users_actions_stats.all_wrong_submissions),3)

        users_actions_stats = users_actions_stats\
                              .fillna(value={'user_success_rate':0})\
                              .set_index('user_id')

        # Добавление данных о последней и предпоследней сессиях студентов
            # Определим временные границы и номера степов последней и предпоследней сессий
        sessions_data_long = days_to_train_submissions_data\
        .sort_values(by=['user_id', 'datetime'],ascending=False)\
        .drop_duplicates(subset=['user_id', 'step_id'])\
        .groupby('user_id',as_index=False)\
        .head(2)

        sessions_data_long['order'] = sessions_data_long\
        .groupby('user_id',as_index=False)\
        .cumcount(ascending=False)-2
            #  order == -1 - последний степ
            #  order == -2 - предпоследний степ

            # Данные последней сессии
        last_session = sessions_data_long\
        .loc[sessions_data_long.order == -1]\
        [['user_id', 'step_id', 'datetime', 'submission_status']]\
        .rename(columns={'step_id':'last_step_id',
                         'datetime':'last_session_end', 
                         'submission_status':'last_submission_status'})

            # Данные предпоследней сессии
        penalt_session = sessions_data_long\
        .loc[sessions_data_long.order == -2]\
        [['user_id', 'step_id', 'datetime',]]\
        .rename(columns={'step_id':'penalt_step_id',
                         'datetime':'penalt_session_end'})

            # Добавление данных о времении окончания предпоследней сессии в submissions_data
        days_to_train_submissions_data_ = days_to_train_submissions_data\
                            .merge(penalt_session[['user_id', 'penalt_session_end']],
                                   how='outer',
                                   on='user_id')

            # Отбор всех ответов каждого студента в последней сессии 
        last_session_data = days_to_train_submissions_data_\
        .loc[(
            (days_to_train_submissions_data_.datetime > days_to_train_submissions_data_.penalt_session_end) | 
            (days_to_train_submissions_data_.penalt_session_end.isnull())
             )]

            # Определение количества ответов в последней сессии для каждого студента
        last_session_data_1 = last_session_data\
        .groupby('user_id', as_index=False)\
        .agg({'datetime':'min', 'step_id':'count'})\
        .rename(columns={'datetime':'last_session_start',
                        'step_id':'last_submissions_number'})

            # Определение количества правильных и неправильных ответов в последней сессии
        last_session_data_2 = pd.pivot_table(last_session_data,
                                            index='user_id',
                                            columns='submission_status',
                                            values='step_id',
                                            aggfunc='count',
                                            fill_value=0)\
        .rename(columns=lambda x: f'last_{x}_submissions')\
        .reset_index()

        last_session = last_session.merge(last_session_data_1,
                                          how='outer',
                                          on='user_id')\
                                   .merge(last_session_data_2,
                                          how='outer',
                                          on='user_id')

            # Определение success_rate (коэффициента успешности каждого студента) в последней сессии
        last_session['last_step_success_rate'] = round(last_session.last_correct_submissions 
                                                       / last_session.last_submissions_number, 3)

        # Слияние данных
        last_session_columns_to_drop = ['last_step_id', 'last_session_end', 'last_session_start', 'last_submission_status']
        users_actions_stats = users_actions_stats\
                              .merge(last_session.drop(columns=last_session_columns_to_drop),
                                     how='outer',
                                     on='user_id')\
                              .fillna(0)

        no_changes_columns_list = ['status', 'entrance_datetime', 'user_success_rate', 'last_step_success_rate']
        columns_list_to_change = [column for column in users_actions_stats.columns if column not in no_changes_columns_list]

        users_actions_stats[columns_list_to_change] = users_actions_stats[columns_list_to_change].astype(int)
        users_actions_stats = users_actions_stats.set_index('user_id')

        self.users_actions_stats = users_actions_stats
    
    
    def get_X_Y_for_ML(self):
        grouped_users_actions_stats = self.users_actions_stats.groupby(self.users_actions_stats.status != 'in progress')
        in_progress_students = grouped_users_actions_stats.get_group(0)
        not_in_progress_students = grouped_users_actions_stats.get_group(1)

        columns_list_to_drop = ['status', 'entrance_datetime']

        self.X = not_in_progress_students.drop(columns=columns_list_to_drop)
        self.Y = not_in_progress_students.status
        
        # В список для прогнозирования не включены студенты, с момента начала обучения которых прошло менее days_to_train дней
        self.X_pred = in_progress_students\
                      .loc[(self.max_events_datetime - in_progress_students.entrance_datetime).dt.days >= self.days_to_train]\
                      .drop(columns=columns_list_to_drop)
```

```python
LM_results = pd.DataFrame()

students_clf = StudentsStatusClassifier()
students_clf.get_drop_out_gap()
students_clf.get_students_status()
students_clf.get_course_completion_time()

# max_days_to_train = min(students_clf.drop_out_gap, students_clf.course_completion_time, 30)
max_days_to_train = students_clf.drop_out_gap + 1

for days_to_train in range(1, max_days_to_train):
    students_clf.get_users_actions_stats(days_to_train)
    students_clf.get_X_Y_for_ML()
    X = students_clf.X
    Y = students_clf.Y
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
    for cross_validation_parts in range(5,10):
#         parameters = {'max_depth': range(1,11),
#                      'min_samples_split': range(2,50,5),
#                      'min_samples_leaf': range(1,20,5)}
        
        parameters = {'max_depth': range(1,20)}

        clf = tree.DecisionTreeClassifier(criterion='entropy')

        search = GridSearchCV(clf, param_grid = parameters, n_jobs=-2, cv=cross_validation_parts)
        search.fit(X_train, Y_train)

        best_tree = search.best_estimator_

        LM_results_temp = pd.DataFrame([search.best_params_], columns = search.best_params_.keys())

        LM_results_temp['cross_validation_parts'] = cross_validation_parts
        LM_results_temp['days_to_train'] = days_to_train

        best_tree_f1_score = f1_score(Y_test, best_tree.predict(X_test), average=None, zero_division=0)
        LM_results_temp['dropped_out_f1_score'] = best_tree_f1_score[0]
        LM_results_temp['passed_course_f1_score'] = best_tree_f1_score[1]
        LM_results = LM_results.append(LM_results_temp)
```
