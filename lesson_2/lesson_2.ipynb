```python
animals_df = pd.read_csv('materials/cats.csv')
animals_df.head()
```

```python
# Перезагрузим датасет, используем содержимое первой колонки для индексации, переименуем названия столбцов.
animals_df = pd.read_csv('materials/cats.csv', index_col=0)\
               .rename(columns={'Шерстист':'furry', 
                                'Гавкает':'barking', 
                                'Лазает по деревьям':'climbing', 
                                'Вид':'animal_type'})

animals_df.head()
```

```python
def ent(df):
    return round(entropy(df.animal_type.value_counts() / df.shape[0], base=2), 2)
```

```python
print(f'For furry == 0: entropy = {ent(animals_df.query("furry == 0"))}')
print(f'For furry == 1: entropy = {ent(animals_df.query("furry == 1"))}')
print(f'For barking == 0: entropy = {ent(animals_df.query("barking == 0"))}')
print(f'For barking == 1: entropy = {ent(animals_df.query("barking == 1"))}')
print(f'For climbing == 0: entropy = {ent(animals_df.query("climbing == 0"))}')
print(f'For climbing == 1: entropy = {ent(animals_df.query("climbing == 1"))}')
```

**Вариант 2:** Кодировка номинативных переменных с помощью функции .cat.codes

##### 1.2.5. Посмотрим как можно уменьшить тендеции к переобучению модели и повысить точность. <br> <br> Ограничим максимальную глубину дерева решений. <br> <br> Также создадим классификатор, обучим его на тренировочных данных.Оценим точность на тренировочных и тестовых данных.

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=5)
clf.fit(X_train, y_train)
print(f'Score based train data: {round(clf.score(X_train, y_train),3)}')
print(f'Score based test data: {round(clf.score(X_test, y_test),3)}')
```

**Точность на тренировочных данных уменьшилась, однако точность классификатора на тестовых данных повысилась. Стоит отметить, что ограничение максимальной глубины дерева решений может также позволить уменьшить время расчётов.**

##### 1.2.6. Попробуем найти оптимальную глубину дерева решений, при которой достигается максимальная точность. Визуализируем результаты расчётов.

```python
max_depth_values = range(1,100)

scores_data = pd.DataFrame()

for max_depth in max_depth_values:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    
    temp_score_data = pd.DataFrame({'max_depth':[max_depth], 
                                    'train_score':[train_score], 
                                    'test_score':[test_score]})
    
    scores_data = scores_data.append(temp_score_data)
    
scores_data\
.reset_index(drop=True)\
.sort_values(by=['test_score'],ascending=False)\
.head(1)
```

```python
# Изменим структуру датафрейма scores_data.
scores_data_long = pd.melt(scores_data, id_vars=['max_depth'],
                          value_vars=['train_score', 'test_score'],
                          var_name='set_type',
                          value_name='score')

scores_data_long.head()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(x='max_depth', y = 'score', hue= 'set_type', data=scores_data_long)
```

##### 1.2.7.Повысить точность модели можно организовав многоуровневую проверку. К примеру, весь датасет разделяется на 3: train dataset - используется для тренировки модели, validation - для её проверки, test - выступает как ещё один уровень проверки. <br> <br> Рассмотрим возможности кроссвалидации, когда тренировочный датасет разделяется на n-е количество частей, одна из которых используется для валидации, а n-1 - для обучения модели. Более подробно о кроссвалидации [здесь](https://scikit-learn.org/stable/modules/cross_validation.html).

- Создадим классификатор;
- Используем ранее полученные данные X_train и y_train для его обучения;
- Выполним оценку точности модели на кроссвалидационном множестве (установите cv=5);
- Попробуем найти оптимальную глубину дерева решений, при которой достигается максимальная точность на кроссвалидационном множестве (реализовать алгорим с кроссвалидацией);
- Визуализируем результаты расчётов.

```python
from sklearn.model_selection import cross_val_score 
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=4)
print(f'Cross-validation score: {round(cross_val_score(clf, X_train, y_train, cv=5).mean(),3)}')
```

```python
max_depth_values = range(1,100)

scores_data = pd.DataFrame()

for max_depth in max_depth_values:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    
    mean_cross_val_score = cross_val_score(clf, X_train, y_train, cv=5).mean()
    temp_score_data = pd.DataFrame({'max_depth':[max_depth], 
                                    'train_score':[train_score], 
                                    'test_score':[test_score],
                                    'cross_val_score':[mean_cross_val_score]})
    
    scores_data = scores_data.append(temp_score_data)

    
scores_data\
.reset_index(drop=True)\
.sort_values(by=['cross_val_score'],ascending=False)\
.head(1)
```

```python
# Изменим структуру датафрейма scores_data.
scores_data_long = pd.melt(scores_data, id_vars=['max_depth'],
                          value_vars=['train_score', 'test_score', 'cross_val_score'],
                          var_name='set_type',
                          value_name='score')

scores_data_long.head()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(x='max_depth', y = 'score', hue= 'set_type', data=scores_data_long)
```

##### 1.4.Попробуем оптимизировать параметры дерева решений
##### 1.4.1. Найдем параметры, при которых достигается максимальная точность на кроссвалидационном множестве с использованием метода `GridSearchCV`.

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Создадим классификатор clf дерева рещений
clf = tree.DecisionTreeClassifier()

# Зададим набор оптимизируемых параметров и диапазоны их варьирования.
parameters= {'criterion':['gini', 'entropy'], 'max_depth':range(1,30)}

# Создадим модель, использующую классификатор clf и "жадные алгоритмы".
grid_search_cv_clf = GridSearchCV(clf, parameters, cv=5)
grid_search_cv_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
print(f'Best classificator precision: {round(precision_score(y_test, best_clf.predict(X_test)),3)}')
print(f'Best classificator recall: {round(recall_score(y_test, best_clf.predict(X_test)),3)}')
```

##### 1.4.3.Определим [вероятность отнесения каждого объекта к определённому классу](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression.predict_proba). <br> <br> Построим гистограмму распределения этих вероятностей.

```python
# Вероятности отнесения к каждому из классов
y_predicted_prob = best_clf.predict_proba(X_test)
# y_predicted_prob
```

```python
# Классы
columns_values = best_clf.classes_
```

```python
# Преобразуем массивы с вероятностями в датафрейм
y_predicted_prob_df = pd.DataFrame(data=y_predicted_prob, columns=columns_values).round(3)

y_predicted_prob_df.head()
```

```python
# Визуализируем данные (распределение вероятностей отнесения пассажиров к классу выживших)
y_predicted_prob_df[1].hist()
plt.xlabel('Probability', fontsize = 12)
plt.ylabel('Frequency', fontsize = 12)
plt.title('Probability of passenger classification as survived', fontsize = 14)
```

```python
# Спрогнозируем класс пассажиров. 
# При этом будет пассажир будет считаться выжившим, если вероятность его отнесения к этому классу >= 0,8.
y_pred = np.where(y_predicted_prob_df[1] > 0.8, 1, 0)
y_pred
```

##### 1.4.4.Построим [ROC-кривую](https://ru.wikipedia.org/wiki/ROC-кривая), которая позволяет оценить качество бинарной классификации. Ссылка на описание из документации scikit-learn [здесь](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_curve.html#sklearn.metrics.roc_curve).

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, y_predicted_prob_df[1])
roc_auc= auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', label='ROC curve (area = %0.3f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
print(f'AUC = {round(roc_auc,3)}.')
if roc_auc > 0.5:
    print('AUC > 0.5. Classificator works in the right way.')
else:
    print('AUC > 0.5. Classificator is wrong.')
```
